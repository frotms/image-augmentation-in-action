```python
import copy
import numpy as np
import Augmentor
import cv2
from matplotlib import pyplot as plt
import time

class AugmAugmentor:
    def __init__(self, cfg, **kwargs):
        self.cfg = cfg
        self.Augmentor = Augmentor
        # # https://github.com/mdbloice/Augmentor/blob/master/notebooks/Multiple-Mask-Augmentation.ipynb
        # # https://augmentor.readthedocs.io/en/master/userguide/examples.html
        # # https://augmentor.readthedocs.io/en/master/userguide/usage.html
        # # https://augmentor.readthedocs.io/en/master/code.html
        # # https://github.com/mdbloice/Augmentor
        # # https://augmentor.readthedocs.io/en/master/userguide/mainfeatures.html#mainfeatures


    def _example(self, image, **kwargs):
        """
        示例函数, 用于示例
        :param image:
        :param kwargs:
        :return:
        """
        p = Augmentor.DataPipeline(images)  # Augmentor.DataPipeline(images, labels)
        p.set_seed(time.time())
        p.skew(probability=1.0, magnitude=1)
        return p.sample(1)


    def aug_image(self, images, **kwargs):
        """
        https://imgaug.readthedocs.io/en/latest/source/examples_basics.html
        images = [
                    [im_1, mask_1_a, mask_1_b],
                    [im_2, mask_2_a, mask_2_b],
                    ...,
                    [im_n, mask_n_a, mask_n_b]
                ]
        :param images:
        :param kwargs:
        :return:
        """
        input_dict = {"images":images} if kwargs.get("labels") is not None else {"images":images, "labels":kwargs.get("labels")}
        p = Augmentor.DataPipeline(**input_dict) # Augmentor.DataPipeline(images, labels)
        p.set_seed(time.time())
        p.random_brightness(probability=1.0, min_factor=0.0, max_factor=1.0)
        p.random_contrast(probability=1.0, min_factor=0.0, max_factor=1.0)
        p.random_distortion(probability=1.0, grid_width=3, grid_height=3, magnitude=1.0)
        p.skew(probability=1.0, magnitude=1)
        #p.status()
        return p.sample(1)
```

```python
augmAugmentor = AugmAugmentor(None)
org_image = cv2.imread("../data/01.jpg")
plt.imshow(org_image)
```

```python
# image
aug_img = augmAugmentor.aug_image([org_image])
plt.imshow(aug_img[0])

print("image augmentation is done!")
```
