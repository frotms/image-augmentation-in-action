```python
import copy
import numpy as np
from matplotlib import pyplot as plt
import mahotas
import time

class MahotasAugmentor:
    def __init__(self, cfg, **kwargs):
        self.cfg = cfg


    def aug_image(self, image, **kwargs):
        image = mahotas.gaussian_filter(image, 1.0, order=0, mode='reflect', cval=0., out=np.empty_like(image).astype(np.float))
        image = mahotas.mean_filter(image, Bc=None, mode='ignore', cval=0.0, out=None)
        image = mahotas.median_filter(image, Bc=None, mode='reflect', cval=0.0, out=np.empty(image.shape, image.dtype))
        image = mahotas.colors.rgb2gray(image, dtype=np.float)

        return image
    
mahotasAugmentor = MahotasAugmentor(None)
org_image  = mahotas.imread("../data/01.jpg")
plt.imshow(org_image)
```

```python
aug_image = mahotasAugmentor.aug_image(org_image)
plt.imshow(aug_image.astype(np.uint8))
```
